# Workshop1 เปรียบเทียบผลของ Hyperparameter
- แสดงให้เห็นถึงความแม่นยำำที่ได้จาก Model ของ Decision Tree เมื่อกำหนดค่าพารามิเตอร์ที่แตกต่างกัน โดยใช้ชุดข้อมูลผู้ป่วยเบาหวาน (เกี่ยวกับข้อมูลดูบทที่ 21) 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
cols = ['preg','plas','pres','skin','insu','bmi','pedi','age','class']
df = pd.read_csv('pima-indians-diabetes.csv',names=cols)
df.info()
```

```python
df.isnull().sum()
```

```python
# check columns ใดบ้างมีค่าเป็น 0 
df['insu'].value_counts()
df.drop(['insu'],axis=1,inplace=True)
df
```

```python
def check_zero(dataframe):
    print('Column 0={} ')
    for column in dataframe.columns:
        zero = dataframe[column][dataframe[column]==0].count()
        zero_portion = zero/(dataframe[column].count())
        print('{} : 0 = {} : {:.2f}% '.format(column,zero,zero_portion*100))
check_zero(df)
```

```python
df.preg.count()
```

```python
cols_replace = ['bmi','skin','pres','plas']
for column in cols_replace:
    df[column] = df[column].replace(0,df[column].median())
df
```

```python
check_zero(df)
```

```python
X = df.drop('class',axis=1)
y = df['class']
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
model1 = DecisionTreeClassifier(criterion= 'gini', max_depth=None)
model2 = DecisionTreeClassifier(criterion= 'entropy',max_depth=None)

models = [model1,model2]
criterion = ['gini','entropy']
for i,model in enumerate(models):
    cvs = cross_val_score(model,X,y,cv=10)
    print('cvs model {}: {}'.format(criterion[i],cvs))
    print('Mean :{:.2f}%'.format(cvs.mean()*100))
```

# Workshop2 เปรียบเทียบAlgorithms 

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
X = X.astype('float64')
X_sc = sc.fit_transform(X)

df_sc = pd.DataFrame(X_sc,columns = X.columns)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
models = []
models.append(('LR',LogisticRegression()))
models.append(('kNN',KNeighborsClassifier()))
models.append(('bayes',GaussianNB()))
models.append(('SVC',SVC()))
models.append(('RFC',RandomForestClassifier()))
models.append(('Tree',DecisionTreeClassifier()))
len(models)
results = []
names = []
for name,model in models:
    cvs = cross_val_score(model,X_sc,y,cv=10)
    results.append(cvs)
    names.append(name)
    print('{:5} {:.3f} ({:.4f}))'.format(name,cvs.mean(),cvs.std()))
```

```python
plt.figure(figsize=(7,7))
plt.rcParams.update({'font.size':14})
ax = plt.subplot()
plt.boxplot(results)
ax.set_xlabel('Algorithms')
ax.set_title('Algorithms Comparison')
ax.set_ylabel('CVS Score')
ax.xaxis.set_ticklabels(names)
plt.show()
```

# Workshop 3 : หาค่าพารามิเตอร์ด้วย Grid Search 

- จาก Workshop3 เท่่่ากับเราได้อัลกอลิทึมแล้ว ลำดับต่อคือการหาว่าค่าพารามิเตอร์ใด ที่ดีจะทำให้อัลกอลิทึม ทำงานแล้วได้ความแม่นยำสูงสุด
    - ใช้ Grid Search ในการหาพารามิเตอร์ที่ดีสุด
    - การทำงานก็คือเรากำหนดค่าต่างๆ ที่ต้องการทดลองรัน ไว้ในตัวแปร Dictionary (ในที่นี้คือ param_grid) ใส่กี่ตัว ตัวละกี่ค่าก็ได้ 
    - โค้ดตัวอย่าง สมมติว่าเราต้องการทดลองค่าพารามิเตอร์ต่างๆ เช่น kernel, C gamma

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C':[1,3,5],
             'kernel':['linear','rbf','poly','sigmoid'],
             'gamma':['auto',1,0.1],
             'shrinking':[True,False],
             'coef0' :[0.0,0.1,0.5]}
model = SVC()
grid_search = GridSearchCV(model, param_grid, cv=10, scoring='accuracy', verbose=10)
grid_search.fit(X_sc,y)
grid_search.best_score_
```

```python
grid_search.best_params_ # ดูพารามิเตอร์ที่ดีที่สุด
```

```python
grid_search.best_estimator_ # ดูพารามิเตอร์ที่ดีที่สุดแบบเต็มๆ
```

```python
print(grid_search.best_estimator_.C)
print(grid_search.best_estimator_.kernel)
print(grid_search.best_estimator_.gamma)
print(grid_search.best_estimator_.shrinking)
print(grid_search.best_estimator_.coef0)
```
