# Workshop3: เตรียมข้อมูลเพื่อ Train และ Test

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.DataFrame(data)
df['target'] = target
df
```

# Using Support Vector Machine Model

```python
# if we use image for data management
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split, cross_val_score
n_samples =len(images)
X = images.reshape(n_samples,-1)
y = target

# let's do Cross Validation
model = SVC(kernel='rbf',gamma='scale')
cvs = cross_val_score(model,X,y,cv= 10)
print(cvs)
print('Mean CVS:', np.array(cvs).mean().round(6)*100,'%')

# Split training and testing data
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size = 0.4,random_state=20)

#Instantiate model and train model
model = SVC(kernel='rbf',gamma='scale')
model.fit(x_train,y_train)

# test model
y_pred = model.predict(x_test)

# evalute model by classification report, confusion matrix
print('Classification Report')
print(classification_report(y_test,y_pred))
y_target = [i for i in range(0,10)]
cm = confusion_matrix(y_test,y_pred)
ax = plt.subplot()
sns.heatmap(cm,annot=True,cmap='YlGnBu')
ax.set_xlabel('Predication')
ax.set_ylabel('Actual')
ax.set_title('Confusion Matrix - Handwritten Number Prediction by SVC Model')
ax.xaxis.set_ticklabels(y_target)
ax.yaxis.set_ticklabels(y_target)
plt.show()
```

# Using Other models -> Naive Bayes, DecisionTreeClassifer, kNN, LogisticRegression

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
import time 
# Instantiate Model for comparison
models = [GaussianNB(), 
          DecisionTreeClassifier(), 
          KNeighborsClassifier(n_neighbors=5),
          SVC(kernel='rbf',gamma='scale'),
          RandomForestClassifier()
         ]
modelNames = ['GaussianNB','DecisionTreeClassifier','kNN','SVC','RandomForestClassifier']
for i,model in enumerate(models):
    t0 = time.time()
    X = data
    y = target
    x_train, x_test, y_train, y_test = train_test_split(X,y, test_size = 0.4)
    print('Model: {}'.format(modelNames[i]))
    print('Cross Validation')
    cvs = cross_val_score(model,X,y)
    print('{} cvs: {}'.format(modelNames[i],cvs))
    print('mean cvs:',np.array(cvs).mean().round(4)*100,'%')
    # Training model
    model.fit(x_train,y_train)
    
    # Test modeling
    y_pred1 = model.predict(x_test)
    
    # Evaluate Model by accuracy_score, Classification Report, Confusion Matrix
    Acc_score = accuracy_score(y_test,y_pred1)
    print("Accuracy_Score: {} %".format(round(Acc_score*100,2)))
    
    
    y_target = [i for i in range(0,10)]
    cm = confusion_matrix(y_test,y_pred1)
    ax = plt.subplot()
    sns.heatmap(cm,annot=True,cmap='YlGnBu')
    ax.set_xlabel('Predication')
    ax.set_ylabel('Actual')
    ax.set_title('Confusion Matrix - Handwritten Number Prediction by {} Model'.format(modelNames[i]))
    ax.xaxis.set_ticklabels(y_target)
    ax.yaxis.set_ticklabels(y_target)
    plt.show()
    t1 = time.time()
    total_time = t1-t0
    
    print('Total time : {} seconds'.format(round(float(total_time),4)))
    print('===========================Done============================')
    print(' ')
    print(' ')
```
