# Chapter 20 : Face Detection and Face Recognition

- Face Detection: เป็นการตรวจว่าภาพใบหน้าคนอยู่ในภาพหรือไม่ ถ้าตรวจเจออาจจะทำการซูมหรือโฟกัส เพื่อนำไปทำ Face Recognition
- Face Recognition: เป็นการตรวจว่าใบหน้าดังกล่าวคือใคร เป็นการรู็จำใบหน้า ก่อนใช้งานจิริงต้องนำใบหน้าของคนเหล่านั้นมา Train บอกคอมพิวเตอร์ ว่านี้คือ นายแดง นางแจ๋ว
- หลักการทำงาน Face Detection
    - มี Algorithms ที่ค่อนข้างซับซ้อนเป้นการคำนวณทางคณิตศาสตร์ ในที่นี้ขอกล่าวเพียง 2 algorithms 
        - 1. Haar Classifier (Face Detection using Haar Feature-basd Cascade Classifier) : เร็ว + ประสิทธิภาพสูง จัดเป็น Supervised ML
            - 1. ทำการสอนก่อนจะใช้งาน โดยนำภาพคนจริงๆ จำนวนมากๆ และภาพที่ไม่มีใบหน้าคน มาสอนให้กับคอมพิวเตอร์ ให้คอมรู้ว่าหน้าคนเป็นอย่างไร และไม่ใช่หน้าคนเป็นแบบไหน
            - 2. ขั้นการใช้งาน เมื่อสอนคอมพิวเตอร์แล้ว เป็นขั้นตอนการตรวจจับ คอมพิวเตอร์ก็จะอ่านข้อมูลไฟล์ภาพนิ่งหรือภาพจากกล้องวิดีโอ จากนั้นก็จะทำการแยกแยะโดยแบ่งซอยแล้วนำตัวเมียบคุณสมบัติเพื่อแยกแยะ features 
        - 2. LBP Classifier (local binary Patterns) 
- หลักการทำงานของ Face Recognition
    - ทำการสอนก่อนนำไปใช้ เพื่อให้รู้ว่าคนนี้คือใคร 
    - ข้อแตกต่างระหว่าง Face Recognition and Face Detection คือ
        - Face Detection นำใบหน้าคนใดก็ได้ สอนว่านี้คือหน้าคน
        - Face Recognition นำรูปใบหน้าหลายๆ รูปของแต่ละคน มาให้ระบบเรียนรู้เช่น รูปนายแดง 40 รูป นายดำ 30 รูปจำนวนต้องมากพอจึงจะแยกแยะได้แม่นยำ เพื่อให้ระบบเรียนรู็และสามารถแยกแยะได้อย่างแม่นยำ
        

```python
%matplotlib inline
from sklearn import datasets
import matplotlib.pyplot as plt
faces = datasets.fetch_olivetti_faces()
faces
```

```python
faces.keys()
data = faces['data']
images = faces['images']
target = faces['target']
DESCR = faces['DESCR']
print(DESCR)
```

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
# df = pd.DataFrame()
df = pd.DataFrame(data) # data = images[i].ravel()
df['target'] = target
def show_image(i):
    '''This function is for showing 12 images, starts with i image'''
    nplots = 50
    plt.figure(figsize = (8,8))
    for j in range(nplots):
        plt.subplot(5,10,j+1) 
        plt.imshow(images[i+j],cmap='gray')
        plt.axis('off')     
        plt.title(target[i+j])
    plt.show()
show_image(50)
```

# Workshop3 เตรียมข้อมูลเพื่อทดลอง Train and Test

- เริ่มจากรูปเพียง 6 รูปก่อน เพื่อให้เข้าใจข้อมูลง่ายขึ้น


```python
n_samples = 60
Xn = images[:60]
Xn.shape
X = Xn.reshape(n_samples,-1)
X
Z = data[:60]
# Z  = X 
y = target[:60]
y
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state = 2)
x_train.shape,x_test.shape
```

```python
# cross val score
model = SVC(kernel='rbf',gamma = 'scale')

cvs = cross_val_score(model,X,y,cv=5)
print(cvs)
print('mean: {}'.format(np.array(cvs).mean().round(4)*100,'%'))

# train model
model.fit(x_train,y_train)

# Test
y_pred = model.predict(x_test)

# Evaluate

y_target = ['Alex','Cala','Dana','Eva','Fox']
print('Classificaton Report')
print(classification_report(y_test,y_pred,target_names=y_target))
cm = confusion_matrix(y_test,y_pred)
ax = plt.subplot()
sns.heatmap(cm,annot=True,cmap='YlGnBu')
ax.set_xlabel('Prediction',color='black')
ax.set_ylabel('Actual')
ax.set_title('Confusion Matrix by SVC model',color='red',fontsize='large')
ax.xaxis.set_ticklabels(y_target,color = 'red')
ax.yaxis.set_ticklabels(y_target)
plt.show()
```
