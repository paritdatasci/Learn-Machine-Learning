# การลดมืตืข้อมูล (Dimensionality Reduction) ลดจำนวนมิติของข้อมูล != ลดขนาดของข้อมูล 
- 2 วิธี 
    - 1. feature selection : เป็นการคัดเลือกหรือกรองข้อมูลบางตัวแปร ออกมาใช้ ซึ่งปัจจุบันมีหลายเทคนิคด้วยกัน
        - Missing Values Ratio : ปกติแล้วเราจะตรวจสอบ Missing Data ก่อน หากมี
            - อาจพิจารณาเติมข้อมูลเข้าไป
            - อาจจะตัดรายการนั้นทิ้ง ตัดแถวนั้นออกไป
            - หากมี Missing Data ประมาณ 40-50% ในcolumns นั้นแสดงว่ามีนัยสำคัญ อาจต้องตัดทิ้งทั้ง columns เลย
        - Low Variance Filter
            - ข้อมูลที่แถบจะมีค่าคงที่ ไม่เปลี่ยนแปลง หรือ low SD ตัวแปรนั้นถือว่าไม่ค่อยมีประโยชน์ อาจตัดทิ้งโดยลบคอลลัมนั้นทิ้งได้
        - High Correlation
            - เป็นการพิจารณาความสัมพันธ์ (Correlation) 
            - พิจารณาว่า which column has high correlation with target/label, if so, we will select such columns for building ML
            - ถ้า column ไหนมี correlation with target ต่ำ อาจจะตัดทิ้งไปเลยก็ได้
            - ลำดับต่อไป พิจารณาเฉพาะ features หากคู่ใดมีความสัมพันธ์กันสูง แปลว่ามีผลต่อ model ในทำนองเดียวกัน ให้เลือกมาเพียงตัวแปรเดียว
    - 2. feature extraction
- ลดขนาดมิติข้อมูลเพื่อให้การ train เร็วขึ้น ใช้ทรัพยากรน้อยลง ลดปัญหา Overfitting
- ลด features != ลด samples (samples ยังคงเท่าเดิม)

# Workshop3 : เลือก Features ด้วย Univariate Selection
- แสดงการเลือก features โดยการทดสอบทางสถิติ เรียกว่า Univariate Selection เป็นการทดสอบสถิติด้วย Chi-Square ซึ่งใช้สำหรับค่า
    - Features ที่มีค่าเป็นบวก  
    - class/target เป็นข้อมูลชนิด Category
    - หลังจากคำนวณแล้วจะได้ค่า Score ของแต่ละคอลลัม จากนนั้นทำการเลือกคอลัม ที่มีค่า Score สูงสุด 
    - ลองไปดูข้อมูลเบาหวาน


```python
# df ได้จาก Workshop ก่อนหน้า ไม่มีคอลลัม insu
df['class'] = df['class'].astype('category')
y = df['class']
X = df.drop(['class'],axis=1)
# หาค่า Chi-Square ของแต่ละ columns
import warnings
from sklearn.preprocessing import MinMaxScaler as Scaler # Normalize ค่าของแต่ละ columns
from sklearn.feature_selection import SelectKBest, chi2
warnings.filterwarnings('ignore')
scaler = Scaler()
X_sc = scaler.fit_transform(X)
df_sc = pd.DataFrame(X_sc,columns = X.columns)
df_sc

# ทำการ import library และ fit ข้อมูลเพื่อเลือก features ที่ดีที่สุด
test = SelectKBest(score_func=chi2,k=5)
fit = test.fit(X_sc,y)

np.set_printoptions(precision=3) # Set ค่าให้ตัวเลขแสดงทศนิยมใน numpy กี่จุดทศนิยม
print(fit.scores_)

features = fit.transform(X_sc)
print(features[0:5,:])
dfts = pd.DataFrame(fit.scores_,columns=['Score'])
dfts['column'] = X.columns
dfts.sort_values(by='Score',ascending=False)
# ถ้าต้องการตัดให้เหลือ 5 คอลลัม ควรตัด Pres and Skin ออก
```

```python
col_fts = dfts.sort_values(by='Score',ascending=False).head(5).column
print(col_fts)
```

```python
X_fts = df_sc.drop(['skin','pres'],axis=1)
print(X_fts.shape)
X_fts
```

นำ 5 columns ไปใช้ต่อในโมเดล เลือกโมเดล SVC ละกัน 

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.svm import SVC
import time
# first do cross validation with 10 k
model = SVC(kernel='linear')
print('With 5 columns!')
t1 = time.time()
cvs = cross_val_score(model,X_fts,y,cv=10)
t2 = time.time()
total_time = t2-t1
print('time Spend:',total_time)

print(cvs)
print('Mean CVS: {}'.format(np.array(cvs).mean().round(4)*100,'%'))
print('========================================')
print('With full columns')

t1 = time.time()
cvs2 = cross_val_score(model,X,y,cv=10)
t2 = time.time()
total_time = t2-t1
print('time Spend:',total_time)
print(cvs)
print('Mean CVS2 :{}'.format(np.array(cvs).mean().round(4)*100,'%'))
```

# การวิเคราะห์องค์ประกอบหลัก PCA

- PCA = Principle Component Analysis (PCA) ซึ่งเป็น Feature Extraction แบบ Unsupervised algorithms ที่นิยมใช้มาก

- PCA หลักการคล้ายกัน คือหามุมมองหรือหาค่ามิติอะไรบางอย่างที่มีมิติจำนวนน้อยที่สุดแต่สามารถใช้เป็นตัวแทนหรือให้ Info ของข้อมูลต้นฉบับได้มากที่สุด หมายถึงกระบวนการ PCA ทำการลดมิติชุดข้อมูลเหลือ 1-2 แล้วทำการ Train ก็ได้ผลลัพธ์เท่าหรือเกือบเท่ากับการใช้ข้มูล 5 columns หรือมากกว่า

- โดยปกติแล้ว ข้อมูลที่มีค่าความแปรปรวน (variance) น้อยมากๆ จะไม่ค่อยมีผลต่อประสิทธิภาพ Model เช่น sample = 1000 และทุกตัวมีค่าเป็น 1 เท่ากันทั้งหมด ลักษณะนี้เรียกว่าความแปรปรวนเป็นศูนย์ เราจะไม่นำมาใช้ พูดง่ายๆ คือเราสนใจข้อมูลที่มีความแปรปรวนสูงๆ 

- ขั้นตอนการทำ PCA: คือการนำข้อมูลไปคำนวณประมวลผลเพื่อสกัดหาข้อมูลที่มีความแปรปรวนสูง โดยจะได้ผลลัพธ์เป็นข้อมูลค่าใหม่ที่อยู่ในตาราง matrix ประะกอบด้วย 
    - PCA จำเป็นต้องใช้ Standard Scaler เพื่อกำจัดปัญหาข้อมูลที่มี Features หรือคอลลัมที่มีช่วงสเกลห่างกันมาก เช่นข้อมูลส่วนสูงกับน้ำหนัก
    -

สังเกตว่า ถ้ามีแกนใด เป็น pca1 รวมอยู่ด้วยนั้นจะสามารถแยกข้อมูลได้ชัดเจนมากยิ่งขึ้น

- ข้อสังเกตคือ pca นั้นจะมีค่าไม่เท่ากัน
- แต่ละค่าเรียกว่า Explained Variance หรือเรียกว่า ค่าอะิบายความแปรปรวนข้อมูล
- PCA มีค่าอยู่ระหว่าง 0-1
- ถ้า PCA ใดมีค่าสูง แสดงว่า PCA นั้นมีความแปรปรวนสูง -> บ่งบอก information ของข้อมูลได้มาก ยิ่งดี ควรนำมาใช้ในการ train model ML
- เราสามารถดู Explained Variance ได้ดังนี้


```python
exp_variance = pca.explained_variance_ratio_
exp_variance
```

- pca1 ค่ามากสุด แปรว่าดีที่สุด คือ 0.73 และ pca2 เท่ากับ 0.229
- ถ้าเลือก pca1 ได้ค่า Explained Variance 0.73 สามารถอธิบายความแปรปรวนหรือมีความสามารถเป็นตัวแทนข้อมูลจริงได้ 73%)  ถ้าเลือก PCA1 และ PCA2 ก็รวมกันได้ 96% (73+23%) 
- ถ้าเลือก pca1 + pca2 + pca3 จะได้ความแปรปรวนรวม กัน 0.73+0.23+0.03 = 99%  
- โดยทั่วไปจะเลือก pca ที่เป็นตัวแทน ได้ 90-95% 
- ในที่นี pca1 + pca2 รวมกันได้ 95% แล้ว เพียงพอ
