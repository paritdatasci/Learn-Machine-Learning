# Missing Data

### ประเภทของ Missing Data
1. ข้อมูลสูญหายแบบสุ่มอย่างสมบูรณ์ (MCAR = missing completely at random)
2. ข้อมูลสูญหายแบบสุ่ม (MAR = missing at random)
3. ข้อมูลสูญหายแบบไม่สุ่ม (MNAR = missing not at random)

# วิธีการจัดการกับ Missing Data

1. นำข้อมูลจริงมาใส่ (ดีที่สุด)
2. การลบข้อมูลที่ไม่สมบูรณ์ออก\
    2.1 Listwise deletion => ลบทั้งแถว สะดวก ไม่ควรเกิน 5% ของข้อมูลทั้งหมดถ้าเราจะลบ\
    2.2 Pairwise deletion => คิดเฉพาะตัวที่มีข้อมูลคือ ไม่ลบ แต่ไม่ทำอะไรกับมัน 
3. การแทนที่ด้วยข้อมูลตัวหนึ่ง (Single imputation Method) เช่น คำนวณค่า Mean/Median จากข้อมูลที่มีแล้วนำมาแทนที่ใน Missing data
4. การแทนที่ด้วยข้อมูลก่อนหน้า หรือ ตำแหน่งถัดไปมาเติม -> มักใช้กับข้อมูลอนุกรมเชิงเวลา เช่น ระบบเก็บอุณหภูมิและความชื้น ปรากฎว่ามีข้อมูล 8.05 หาย จึงใช้ข้อมูล 8.00 มาแทน
5. การคำนวณทางสถิติ -> ใช้การวิเคราะห์ด้วยเทคนิคต่างๆ แล้วนำค่าที่ได้มาเติม เช่นใช้ regression นำค่า X ที่มีไปคำนวณหาค่า Y ในสมการถดถอย แล้วนำไปเติมใน Missing data 
6. Multiple Imputation
7. Kth Nearest Neighbor (KNN) imputation
8. Iterative Model-Based Imputation
9. Random Forest Imputation

# Workshop 14 ลบแบบมีเงื่อนไข เช่นกำหนดว่าต้องมีข้อมูลเหลือกี่ตัวขึ้นไป (ที่ไม่เป็น missing data) ถ้าเหลือน้อยกว่านี้จะลบคอลลัมทิ้ง เช่นกำหนด Threshold ไว้ 8 ถ้าในคอลลัมใด มีข้อมูลเหลือน้อยกว่า 8 มันจะลบคอลลัมทิ้ง

```python
df = pd.read_csv('data_missing.csv')
df
```

```python
df.dropna(thresh= 8,axis=1,inplace =  True)
```

```python
df # พบว่า columns Temp ถูกลบออกเพราะมี missing value และ ข้อมูลที่เหลือน้อยกว่า 8 ตัว
```

### กำหนดสัดส่วนเงื่อนไขคือ ถ้าข้อมูลปกติเหลอไม่ถึง 80 % ของจำนวนทั้งหมด ให้ลบคอลลัมทิ้งไป พูดง่ายๆคือ มี missing value > 20 %

```python
df.shape[0]
```

```python
left = int(df.shape[0]*.8)
left
```

```python
df.dropna(thresh = left, axis =1 ,inplace=True )
```
