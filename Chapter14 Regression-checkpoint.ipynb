# สรุป Summary Regression(สมการการถดถอย)
- Regression เป็นการศึกษาความสัมพันธ์ของตัวแปร 2 ตัวขึ้นไป ส่วนใหญ่ใช้ x,y แทนตัวแปร ในลักษณะถ้าค่าตัวแปรหนึ่งเพิ่มจะส่งผลให้ค่าตัวแปรเปลี่ยนแปลงไปด้วย ซึ่งจะมีความสัมพันธ์แบบเชิงเส้นตรง หรือเส้นโค้งในลักษณะที่เพิ่มขึ้นหรือลดลงก็ได้

```python
import numpy as np
y = np.array([100,50,30,20])
y_predict= np.array([90,50,50,30])

MAE = (abs(y - y_predict)).sum()/len(y)
print(MAE)

from sklearn.metrics import mean_squared_error, mean_absolute_error
print(mean_absolute_error(y,y_predict))
```

ตัวอย่าง MSE

```python
print(mean_squared_error(y,y_predict))
```

ตัวอย่าง RMSE

```python
np.sqrt(mean_squared_error(y,y_predict))
```

### หลักการโดยรวมคือการหาระยะคลาดเคลื่อนของจุดจริงเทียบกับ Model
- MAE เป็นค่าที่เข้าใจได้ง่ายเพราะ มันคือค่าเฉลี่ยของระยะคาดเคลื่อน Average Error
- RMSE เป็นค่าที่ได้รับความนิยมมากกว่า เนื่องจากสามารถแปลความหมายเป็นหน่วยของy ได้

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline 
from sklearn.linear_model import LinearRegression

model = LinearRegression()
x = [1.0,1.8,3.0,4.1,5.2,6.0]
y = [1,1.3,2.2,2.5,2.8,3.6]

x = np.array(x).reshape(-1,1)
y = np.array(y).reshape(-1,1)

model.fit(x,y)
prd = model.predict(x)
sns.set_style('darkgrid')
plt.scatter(x,y)
plt.plot(x,prd)
plt.scatter(x,prd,marker='s',c='r')
plt.show()

# Evaluate Model

from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score

print('Mean Squared Error(MSE) : {:.4f}'.format(mean_squared_error(y,prd)))
print('Mean Absolute Error(MAE) : {:.4f}'.format(mean_absolute_error(y,prd)))
print('Coefficient of Determination (R square) : {:.4f}'.format(model.score(x,y)))

```

# การหาค่า R2 score

```python
model.score(x,y) # ค่า R2 ที่ได้
```

- R square (coefficient of Determination) : ความหมายคือ Model มีความถูกต้องแม่นยำประมาณ 97 % พูดง่ายๆ คือ ถ้านำ Model ไปทำนาย Predict หรือประมาณการจะได้ Output ที่ถูกต้องประมาณ 97%
- MSE and MAE : เท่ากับ 0.024 และ 0.1242 ตามลำดับ ซึ่งใกล้เคียง 0 แปลว่า มีความคาดเคลื่อนต่ำ

# Workshop 7 ทำนาย GDP ด้วย Polynomial Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

df = pd.read_csv('gdp4c.csv',index_col=0)
x = df.index.values.reshape(-1,1)
y = df.THA
poly_features = PolynomialFeatures(degree=2)
x_poly = poly_features.fit_transform(x)
model = LinearRegression()
model.fit(x_poly,y)
y_predict = model.predict(x_poly)
```

```python
plt.scatter(x,y,label='Real Data')
plt.plot(x,y_predict,c='r',label='Polynomial Regression degree=2')
plt.title('GDP-THAILAND 2 degree Model')
plt.legend()
```

# ประเมินความแม่นยำของโมเดล MSE,RMSE,R2

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
MSE_2d = mean_squared_error(y,y_predict)
RMSE_2d = np.sqrt(MSE)
r2_2d = model.score(x_poly,y)
print(MSE_2d)
print(RMSE_2d)
print(r2_2d)
```

# Overfitting และ Underfitting
- Overfitting คือ ใช้ Degree สูงไปเวลาเอาไปทำนายมันไม่ได้เป็นเช่นนั้น
- Underfitting คือ ใช้ Degree ต่ำไปเช้นความสัมพันธ์เป็นเส้นโค้งแต่เราใช้เส้นตรง จึงไม่เหมาะสม
- ปกติแล้วการเลือก Degree เราจะทำหลายๆ Degree ค่อยๆไล่และดูกราฟประกอบกันไป

# ลอง Covid 19 ด้วย Degree = 4

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

df = pd.read_csv('covid19.csv')
df = df[df.index>65]
x = df.index.values.reshape(-1,1)
y = df.newcase

poly_features = PolynomialFeatures(degree=4)
x_poly = poly_features.fit_transform(x)

model = LinearRegression()
model.fit(x_poly,y)

y_prd = model.predict(x_poly)

plt.scatter(x,y)
plt.plot(x,y_prd,c='r',linestyle=':',label='prediction')
plt.legend()
plt.show()

print('MSE',mean_squared_error(y,y_prd))
print('RMSE',np.sqrt(mean_squared_error(y,y_prd)))
print('r2',r2_score(y,y_prd))
```
