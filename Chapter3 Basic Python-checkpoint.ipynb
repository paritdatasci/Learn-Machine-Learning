# Function

```python
area = lambda w,h:w *h
area(5,7)
```

```python
def area(w,h):
    area = w * h
    return area
area(7,8)
```

```python
def area(length):
    return length**2
if __name__ =='__main__':
    b = area(4)
    print('Area = %d'%(b))
    print(__name__)
    print(type(__name__))
```
