# Workshop39 จัดกลุ่มด้วย CUT

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'Age':np.random.randint(12,55,20)}) # generate age between 12:55 for 20 items
labels =['teen','old']# แบ่งกลุ่มอายุ 
groups = [12,20,60] # แบ่งอายุคนตั้งแต่ 12-20,20-60
df['stages'] = pd.cut(df['Age'],bins=groups,labels = labels)
df
```

```python
import pandas as pd
import numpy as np
df_salary = pd.DataFrame({'Salary':np.random.randint(10000,90000,100)})
labels = ['Low','Mid','High']
groups = [10000,30000,60000,90000]
df_salary['labels'] = pd.cut(df_salary['Salary'],bins=groups,labels = labels)
df = df_salary
df
```

```python
df['Age'] = pd.Series(np.random.randint(20,55,100))
df
```
