# ตอนที่ 20 Confusion Matrix บอกว่าผลลัพธ์ที่เราทำนายผิดถูกตรงไหนบ้าง

```python
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_predict # การทำนายผลจากการทดลอง แต่ละครั้ง ตามจำนวน k-fold
def displayConfusionMatrix(cm,cmap=plt.cm.GnBu):
    classes=["Other Number","Number 5"]
    plt.imshow(cm,interpolation='nearest',cmap=cmap)
    plt.title("Confusion Matrix")
    plt.colorbar()
    trick_marks=np.arange(len(classes))
    plt.xticks(trick_marks,classes)
    plt.yticks(trick_marks,classes)
    thresh=cm.max()/2
    for i , j in itertools.product(range(cm.shape[0]),range(cm.shape[1])):
        plt.text(j,i,format(cm[i,j],'d'),
        horizontalalignment='center',
        color='white' if cm[i,j]>thresh else 'black')

    plt.tight_layout()
    plt.ylabel('Actually')
    plt.xlabel('Prediction')
    plt.show()
y_train_pred = cross_val_predict(sgd_clf,x_train,y_train_5,cv=3)
cm = confusion_matrix(y_train_5,y_train_pred)
print(cm)
displayConfusionMatrix(cm)
```

# ตอน 24 ทำนายโรคเบาหวานด้วย KNN 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
df = pd.read_csv('https://raw.githubusercontent.com/kongruksiamza/MachineLearning/master/KNN/diabetes.csv')
x = df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']]
x = x.values
y = df[['Outcome']]
print(x)
y = y.Outcome.values
# print(df)
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.4)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
# Create Model and Train Model

k_neighbors = np.arange(1,9)

training_score = np.empty(len(k_neighbors))
testing_score = np.empty(len(k_neighbors))

for i,k in enumerate(k_neighbors):
    #1,2,3,4,5,6,7,8,9
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(x_train,y_train)
    # วัดประสิทธิภาพโมเดล โดยเกิดจากกำหนดค่า k ที่แตกต่างกัน เลือกค่าที่เหมาะสม
    training_score[i] = knn.score(x_train,y_train)
    testing_score[i] = knn.score(x_test,y_test)
    print(testing_score[i]*100)
plt.title('Compare k value in Model')
plt.plot(k_neighbors,testing_score,label='Test Score')
plt.plot(k_neighbors,training_score,label='Train Score')
plt.xlabel('K number')
plt.ylabel('Score')
plt.legend()
plt.show()
# เอาค่า test score ไปใช้ 
```

# ตอนที่ 28 ทำนายรายได้ประชากรด้วย GaussianNB


```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

def cleandata(dataset):
    for column in dataset.columns:
        if dataset[column].dtype == type(object):
            lb = LabelEncoder()
            dataset[column] = lb.fit_transform(dataset[column])
    return dataset

def split_feature_class(dataset,feature):
    features = dataset.drop(feature,axis=1)
    label = dataset[feature].copy()
    return features,label


df = pd.read_csv('adult.csv')
dataset = cleandata(df)


# train_set, test_set
train_set, test_set = train_test_split(dataset,test_size=0.2)

# train_set 
train_feature, train_label = split_feature_class(train_set,'income')

# test_set
test_feature, test_label = split_feature_class(test_set,'income')

# Create Model
model = GaussianNB()
model.fit(train_feature,train_label)

NB_pred = model.predict(test_feature)

print('Accuracy Score: {}'.format(accuracy_score(test_label,NB_pred)))
```

Accuracy Score ไม่สูงนักเพราะข้อมูลเป็น NUll เยอะพอสมควร NB ที่ดีควรมี Acc Score roughly 85% up!!!

# การวิเคราะห์องค์ประกอบหลัก (Principal Component Analysis: PCA)


วิเคราะข้อมูลหลายตัว เพื่อหาความสัมพันธ์เหล่านั้น เช่น เรามีตัวแปรทั้งหมด 50 ตัว อาจจะลดขนาดลงมาเหลือไม่กี่ตัว เช่น วิเคราะห์ว่าตัวแปรใดบ้างที่่มีความสัมพันธ์กับตัวแปร ตัวใดไม่มี เวลาเลือกไปใช้งานมันจะสร้าง model ดีขึ้น  เร็วขึ้น แม่นยำขึ้น

PCA จะทำการสร้างตัวแปรที่เรียกว่า component โดยแต่ละ Component จะไม่มี ความสัมพันธ์กันเลย component ตัวแรก จะมีค่า variance สูงสด ซึ่งจะอธิ
จะอธิบายข้อมูลได้มากที่สุด แล้วตัวถัดไป variance จะลดลงตามลำดับ

```python
from sklearn.datasets import make_blobs # กำหนดได้ว่าจะมีกี่ features
from sklearn.decomposition import PCA # เพื่อลดจำนวน feature เพื่อให้เลือก feature ที่เราได้กำหนดเอาไว้
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
x,y = make_blobs(n_samples=100,n_features=10)
# x เก็บข้อมูล sample 
# print(x) # มีข้อมูล 100 แถว 10 คอลลัม
print('Before',x.shape) # before
# print(y.shape) # เก็บข้อมูล 100 แถว ตัว 
# ตัว PCA ใช้เพื่อลด features

pca = PCA(n_components = 4)
pca.fit(x)
x = pca.transform(x) # แปรงกับ
# x = pca.fit_transform(x)# แบบนี้ก็ได้
print('After',x.shape)


print(pca.n_components_) # ดูว่ามีกี่ components
print(pca.explained_variance_ratio_) # ค่าvariance ตัวแรกเยอะสุด อยู่ที่ 0.625 คือมีผลต่อข้อมูลเยอะสุด

df = pd.DataFrame({'variance':pca.explained_variance_ratio_,'pc':['PC1','PC2','PC3','PC4']})
sns.barplot(x='pc',y='variance',data=df,color='c')
plt.show()
```

# USE PCA with real dataset IRIS Compared with doing model without PCA 

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

iris = sns.load_dataset('iris')
iris

x = iris.drop('species',axis= 1)
y = iris['species']
print('Before',x.shape)
# เอา x ไปใช้ใน PCA เช่น ลดให้เหลือ 3 attribute
pca = PCA(n_components = 3)
x_pca = pca.fit_transform(x)
print('After',x.shape)

print(pca.explained_variance_ratio_) # ค่าvariance ตัวแรกเยอะสุด อยู่ที่ 0.625 คือมีผลต่อข้อมูลเยอะสุด
# add before,after with PCA components into dataframe x
x['PCA1'] = x_pca[:,0]
x['PCA2'] = x_pca[:,1]
x['PCA3'] = x_pca[:,2]
x # ใช้ ข้อมูลเพียง columns PCA ไปทำ model เท่านั้น
# load_data

x_train,x_test,y_train,y_test = train_test_split(x,y)
# Complete data
x_train = x_train.loc[:,['PCA1','PCA2','PCA3']]
x_test = x_test.loc[:,['PCA1','PCA2','PCA3']]

# create model and train model
model = GaussianNB()
model.fit(x_train,y_train)
# predict model
y_pred = model.predict(x_test)
print('Accuaracy Score', accuracy_score(y_test,y_pred))
```

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
x,y = make_blobs(n_samples=300,centers = 4,cluster_std = 0.5,random_state=0) # center คือการแบ่งข้อมูลเป็นกี่กลุ่ม
x # เก็บขนาด และ columns (features)
print(x.shape)
# print(x[:,0])
# print(y) # เก็บ label
# check scatterplot of data
plt.figure(figsize=(8,6))
plt.scatter(x[:,0],x[:,1]) # x[:,0] -> แกนx , x[:,1] -> แกน y
# ใช้ตัว k -means เป็นการค้นหา 
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=4)
model.fit(x) # train data พร้อมระบุค่า centroid
y_pred = model.predict(x)
centers = model.cluster_centers_



x_test,y_test = make_blobs(n_samples = 10, centers= 4,cluster_std= 0.5,random_state=0)

y_pred_new = model.predict(x_test)




plt.figure(figsize=(8,6))
plt.scatter(x[:,0],x[:,1],c=y_pred)
plt.scatter(x_test[:,0],x_test[:,1],marker='s',c=y_pred_new,s=120)
plt.scatter(centers[0,0],centers[0,1],alpha=0.2,c='r',marker='x',label='Centroid1')
plt.scatter(centers[1,0],centers[1,1],alpha=0.2,c='salmon',marker='x',label='Centroid2')
plt.scatter(centers[2,0],centers[2,1],c='g',marker='x',alpha=0.2,label='Centroid3')
plt.scatter(centers[3,0],centers[3,1],c='black',marker='x',alpha=0.2,label=('Centroid4'))
plt.legend(frameon=True)
plt.show()
print(centers)


# โยนข้อมูลชุดทดสอบ สมมติ สร้าง กลุ่มข้อมูลใหม่
# new point
```

#### ใช้วิธีการ PCA (Principal Component Analysis) เพื่อลดขนาดของ Features และนำไปทำงานร่วมกับ SVM (Support Vector Machine) เพื่อเรียนรู้และสร้างโมเดลทำนายผลใบหน้า 
ปัจจุบัน SVM โดนถูกแทนที่ด้วย Neural Network ซึ่งทำงานได้ดีกว่า SVM 

```python
from sklearn.datasets import fetch_lfw_people
# download and display image
faces = fetch_lfw_people(min_faces_per_person=60)
print(faces.target_names) # ชื่อเจ้าของใบหน้า
```

```python
print(faces.images.shape)
```

```python
import matplotlib.pyplot as plt

print(faces.keys())
print(faces.target_names)
```

```python
# plt.imshow(faces.images[0],cmap='gray',interpolation='nearest')
# for i in range(15):
fig,ax = plt.subplots(3,5)
for i, axi in enumerate(ax.flat):
    axi.imshow(faces.images[i],cmap='bone')
    axi.set(xticks=[],yticks=[])
    axi.set_ylabel(faces.target_names[faces.target[i]].split()[-1],color='black')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline # เชื่อมต่อ pca กับ svc
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
# Reduce & Create model
pca = PCA(n_components=150,svd_solver='randomized',whiten = True) # ต้องสร้างส่วนที่เชื่อมตัว
svc = SVC(kernel='rbf',class_weight = 'balanced')
model = make_pipeline(pca,svc)
# train, test data
x_train, x_test, y_train, y_test = train_test_split(faces.data,faces.target,random_state=40)



# สร้าง dict เก็บค่า c หรือ gamma 
param = {'svc__C':[1,5,10,50],
         'svc__gamma':[0.0001,0.0005,0.001,0.005]}
# train data to model
grid = GridSearchCV(model,param)
grid.fit(x_train,y_train)
print(grid.best_params_)


```
