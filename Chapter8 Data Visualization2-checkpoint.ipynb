### TA TB TC คืออัตราการให้ปุ๋ยของ 3 ชนิด เช่น TA =20 คือการให้ปุ๋ย TA 20 กรัม

# สรุป :บทนี้ เน้นศึกษาแนวโน้มความสัมพันธ์ของตัวแปร โดยใช้ Data Visualization รูปแบบต่างๆ ซึ่งเป็นขั้นตอนสำคัญก่อนจะไปศึกษา Machine learning

### scatterplot: สำหรับพลอตจุดข้อมูล เพื่อดูการกระจาย ดูแนวโน้มและความสัมพันธ์

### jointplot: สำหรับพล็อต คล้าย scatterplot แต่มีการเพิ่มกราฟการกระจายกำกับของทั้งสองแกนประกอบด้วย มี Histogram ของแต่ละแกน
### swarmplot: พล็อตดูการกระจาย เหมาะสำหรับค่าแกน x ที่เป็นข้อมูลชนิดไม่ต่อเนื่ง
### Correlaton: ค่าแสดงความสัมพันธ์ของสองตัวแปร
### Heatmap : เหมาะสำหรับดูความสัมพันธ์ระหว่างตัวแปร ในรูปแบบค่าสัมประสิทธิ์สหสัมพันธ์ เข้มมากแปรว่า corr มาก
### pairplot: สำหรับสั่งพล็อตทีเดียวเพื่อดูทุกความสัมพันธ์ระหว่างกันในตัวแปรใน dataframe
### lmplot: สำหรับดูเส้นแนวโน้มความสัมพันธ์แบบเชิงเส้น linear regression

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_csv('treatment-nl.csv')
df.sample(5)
```

# Workshop 2 พลอตดูการกระจาย

```python
sns.set_style('darkgrid')
sns.distplot(df['product'],color='g')
```

```python
df.corr()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.title('TA')
sns.scatterplot(x='TA',y='product',data=df) # เมื่อ TA สูงขึ้น product ก็สูงขึ้น


plt.subplot(1,3,2)
plt.title('TB')
sns.scatterplot(x='TB',y='product',data=df)


plt.subplot(1,3,3)
plt.title('TC')
sns.scatterplot(x='TC',y='product',data=df)
```

```python
df.corr(method='pearson')
```

### เห็นความสัมพันธ์ว่าการให้ปุ๋ย TA ค่อนข้างส่งผลดีต่อผลผลิต ในขณะที่ TB และ TC แถบจะไม่มีผลต่อผลผลิต

```python
sns.scatterplot(x='TA',y='product',data=df,label='TA')
sns.scatterplot(x='TB',y='product',data=df,label='TB')
sns.scatterplot(x='TC',y='product',data=df,label='TC')
plt.legend(loc='lower right')
```
