# Workshop 10 : กำหนดรายละเอียด

```python
xs = [1,2,3,4,5,6]
ydata = [1,2.8,2.1,4,3,3.5]
ydata2 = [2.2,1.6,2.3,2,2.3,2.1]

plt.rcParams.update({'font.size':13, 'font.family':'tahoma'})
plt.figure(figsize = (8,6))

plt.grid(linestyle='--', axis='y') # เส้นกริด เฉพาะแกน y
plt.bar(xs,ydata,color='c', width=0.5, label= 'Domestic')
plt.bar([x + 0.15 for x in xs],ydata2, color='orange', width=0.5,label='Export',alpha = 0.7)


plt.plot(xs,ydata,color='b',marker='s',markersize=8)
plt.plot([x + 0.15 for x in xs], ydata2,color='r',marker='o',markersize=8)

plt.title('รานงาน Sales Report')
plt.xlabel('Month')
plt.ylabel('Sales')

plt.ylim(0,4.5)


plt.legend()
plt.show()
```

```python
xs = [1,2,3,4,5,6]
ydata = [1,2.8,2.1,4,3,3.5]
ydata2 = [2.2,1.6,2.3,2,2.3,2.1]

import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.bar(xs,ydata,color='c', width= 0.5, label = 'Domestic Sales')
plt.bar([x + 0.2 for x in xs],ydata2, color='orange', width = 0.5,alpha = 0.4, label= 'Export Sales')


plt.plot(xs,ydata, color='blue',marker='s',markersize=5)
plt.plot([x + 0.2 for x in xs],ydata2, color='red',marker='o',markersize=5)

plt.ylabel('Sales')
plt.xlabel('Month')
plt.title('Sales Report')

plt.legend()
```
