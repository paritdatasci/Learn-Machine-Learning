# Workshop6 การหาค่าเฉลี่ยและการรวม
กล่าวคือ เมื่อมีข้อมูลจำนวนมาก โดยเฉพาะอย่างยิ่งข้อมูลมีการขึ้นลงดังตัวอย่าง เส้นกราฟอาจจะแน่นจนดูไม่ออก หากต้องการดูแนวโน้ม ค่ากลาง จะต้องใช้วิธีหาค่าเฉลี่ย  โดยมีวิธีดังนี้
1. Rolling : นำข้อมูลในช่วงเวลาหนึ่งๆมาหาค่าเฉลี่ย เช่น ข้อมูลในตัวอย่างนี้เก็บค่าแบบรายวัน ต้องการหาค่าเฉลี่ยรายสัปดาห์ ก็นำค่า 7 วันมาเฉลี่ยแล้วแทนค่าที่ศูนย์กลาง(center)
2. Resampling: นำข้อมูลในช่วงเวลาหนึ่งมาหาค่าเฉลี่ยเหมือนกัน แต่จำนวนข้อมูลจะเพิ่มลดตามที่กำหนด เช่น จาก 7 วัน (7แถว) พอเฉลี่ยจะลดเหลือ 1 แถว -> พูดง่ายๆ คือ 7 rows ยุบเหลือ 1  Rows

```python
df.resample('7d').mean()
```

```python
df.rolling(7,center=True).mean()
```

```python
df.rolling(7,center=True).sum()
```

# Workshop8 พลอตกราฟ Resample and Rolling

```python
start,end = '2017-1','2017-12'
plt.figure(figsize=(15,4))
plt.rcParams.update({'font.size':18})
df.Solar.loc[start:end].plot(marker='o',label='Daily',markersize=2,alpha=0.7)
df.Solar.loc[start:end].resample('15d').mean().plot(marker='o',c='b',markersize=8,label='Mean Sampling')
df.Solar.loc[start:end].rolling(15,center=True).mean().plot(marker='o',markersize=4,label='Mean Rolling')

plt.ylabel('Power GWh')
plt.title('Solar')
plt.legend()
plt.grid()
plt.show()
```

### ข้อสังเกต กราฟของข้อมูลจริงมียอดแหลม เนื่องจากมีการบันทึกค่ารายวัน ข้อมูลมีการแกว่งมาก
### Rolling and Resample ช่วยทำการเฉลี่ยข้อมูล ทำให้ยอดแหลมลดลง เหมาะสำหรับการดูแนวโน้ม หรือเป็นการลด Noise
### Resample เส้นจะมีลักษณะเป็นหยักๆเหลี่ยมๆมากกว่า Rolling เนื่องจากมีการลดจำนวนแถวข้อมูล ถูกยุบลดลงตามที่ระบุ

# Workshop 10 ข้อมูล Time Series ที่มีค่าเวลา

```python
import pandas as pd
import numpy as np
range = pd.date_range('2020-01-1','2020-02-28',freq='10min')
```

```python
df = pd.DataFrame(index=range)
```

```python
df['Temp'] = np.random.randint(10,40, size=df.shape[0])
df['Humid'] = np.random.randint(50,70,size=df.shape[0])
df.reset_index(inplace=True)
df
```

```python
df.columns = ['Time','Temp','Humid']
df
```

```python
df.Time = pd.to_datetime(df.Time)
```

```python
df.set_index('Time',inplace=True)
df
```

```python
df.loc['2020-1-15 09:00:00':'2020-1-15 09:30:00']
```

```python
start = pd.to_datetime('2020-1-15 10:00')
end = pd.to_datetime('2020-1-18 1:00')
ax = df.loc[start:end].plot(grid=True,figsize=(11,4),subplots=True,sharex=False)
plt.show()
```

# สามารถใช้ resample หรือ rolling หาค่าเฉลี่ยได้ เช่นต้องการกราฟเฉลี่ยรายชั่วโมง

```python
df.resample('1h').mean().loc[start:end].plot(grid=True)
```

```python
df.rolling(6,center=True).mean().loc[start:end].plot() # ใช้เลข 6 เพราะ roll 6 ครั้งเท่ากับ 1 ชั่วโมง
```

```python
df_temp = df.Temp.loc[start:end]
print(start,end)
df_resampling = df_temp.resample('1h').mean()
df_rolling = df_temp.rolling(6,center=True).mean()

plt.figure(figsize=(14,5))
df_temp.plot(grid=True,alpha=0.5,label='raw data',rot=20)
df_resampling.plot(grid=True,marker='s',c='b',label='resample',rot=20)
df_rolling.plot(grid=True,marker='o',c='m',markersize=0.9,label='rolling',rot=20)
plt.xlabel('Date time')
plt.ylabel('Temperature')
plt.title('Temperature from Start to Ends')
plt.legend()
plt.ylim(5,50)
plt.show()
```

```python
def convert_date(date):
    date = date.split('/')
    year = date[2]
    year = str(int(year) - 543)
    
    date = '-'.join([year,date[1],date[0]])
    return date
df['Date'] = df['Date'].apply(convert_date)
df
```
